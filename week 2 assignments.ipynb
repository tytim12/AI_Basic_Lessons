### 课堂作业

计算闰年  

```python
# 闰年计算规则
# 能被4整除的，同时能被 100和400 整除
# 能被4整除，但不能被100整除

try:
    year = int(input("请输入年份："))
except:
    year = int(input("请正确输入年份："))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("{}年是闰年！".format(year))
        else: print("{}年不是闰年！".format(year))
    else: print("{}年是闰年！".format(year))
else: print("{}年不是闰年！".format(year))
```

### 作业1： 实现一周的记账功能
 - 首先记录输入的7天的收入
 - 然后记录输入的7天支出
 - 然后打印以下内容：
 > 1. 7天的收入（list）
 > 2. 7天的支出（list）
 > 3. 每天的结余（dict）
 > 4. 总结余  
 
 模拟数据  
 income = [200, 210, 180, 195, 220, 240, 200]  
 cost = [180, 200, 100, 50, 100, 140, 80]

```python
# 预设数据版本
income = [200, 210, 180, 195, 220, 240, 200]
cost = [180, 200, 100, 50, 100, 140, 80]
daily_balance = {}

# 每日结余
for i in range(len(income)):
    key = "周"+ str(i+1) + "结余"
    daily_balance[key] = int(income[i]) - int(cost[i])
    print("{}{}元".format(key, daily_balance[key]))

print("总结余 = {}元".format(sum(income) - sum(cost)))
```

```python
# 自行输入版本 - 一天一天输入
income = []
cost = []
daily_balance = {}

for day in range(3):
    d_income = int(input("第{}天收入：".format(day+1)))
    d_cost = int(input("第{}天支出：".format(day+1)))
    income.append(d_income)
    cost.append(d_cost)
  
# 每日结余
print("\n每天结余：")  
for i in range(len(income)):
    key = "周" + str(i+1) + "结余"
    daily_balance[key] = income[i] - cost[i]
    print("{}{}元".format(key, daily_balance[key]))
    
print("总结余 = {}元".format(sum(income) - sum(cost)))
```

### 作业2：根据操作查询余额，存款，取款和退出
==========  
请选择操作  
1. 查询余额  
2. 存款  
3. 取款  
4. 退出  

==========

```python
# 初始选项

savings = 0

while True:
    
    print("\n")
    print("============")
    print("请选择操作")
    print("1. 查询余额")
    print("2. 存款")
    print("3. 取款")
    print("4. 退出")
    print("============")
    print("\n")
    
    choice = int(input("请输入选择："))

    if choice == 1:
        print("余额为：", savings)
    
    elif choice == 2:
        deposite = float(input("请输入存款金额："))
        savings += deposite
        print("已存款：", deposite, "元")
    
    elif choice == 3:
        withdraw = float(input("请输入取款金额："))
        if withdraw > savings:
            print("余额不足")
        else:
            savings -= withdraw
            print("已取款：", withdraw, "元")
    
    elif choice == 4:
        break
    else: pass
        
        
```
